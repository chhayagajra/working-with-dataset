```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline            # a histogram plot to check the frequency distribution of the variable ‘mpg’ (Miles per gallon).
plt.hist(df['mpg'],color='green',edgecolor='white',bins=5) 

plt.xlabel ('mpg1')
plt.ylabel ('frequency')
plt.title ('histogram of miles per gallons')

plt.show() 
```

```python
plt.scatter(df['wt'],df['mpg'],c='red') #data visualisation scatter plot.
plt.title('scatter plot of mpg v/s weight of cars')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

It can be inferred from scatter plot of ‘mpg’ (Miles per gallon) vs ‘wt’ (Weight of car) from the dataset mtcars1.csv that as the weight of the cars increases the mpg decreases.
